```python
# 1. Импорты и базовые настройки

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score,
    classification_report,
    mean_absolute_error, mean_squared_error, r2_score
)

plt.rcParams["figure.figsize"] = (8, 5)
pd.set_option("display.max_columns", 50)

# 2. Загрузка и предобработка данных

# Загрузка датафрейма
df = pd.read_csv("ad_performance_results.csv")

# Если некоторые строки в названиях Audience содержат лишние пробелы, еберем их
df["Audience"] = df["Audience"].astype(str).str.strip()

# Посмотрим на первые строки датафрейма, чтобы понять структуру данных.
print("Первые строки исходного датафрейма:")
display(df.head())
print("\nИнформация о датафрейме:")
df.info()

# Функции для приведения строковых чисел к числам

def parse_int(x):
    if pd.isna(x):
        return np.nan
    s = str(x).replace(",", "").strip()
    if s == "":
        return np.nan
    try:
        return int(float(s))
    except ValueError:
        return np.nan

def parse_float(x):
    if pd.isna(x):
        return np.nan
    s = str(x).replace(",", "").strip()
    if s == "":
        return np.nan
    try:
        return float(s)
    except ValueError:
        return np.nan

# Приводим метрики к числам
df["Impr"] = df["Impr."].apply(parse_int)
df["Clicks_num"] = df["Clicks"].apply(parse_int)
df["Conversions_num"] = df["Conversions"].apply(parse_int)
df["Cost_num"] = df["Cost"].apply(parse_float)

# Пересчитываем метрики
df["ctr_calc"] = np.where(df["Impr"] > 0,
                          df["Clicks_num"] / df["Impr"],
                          0)

df["conversion_rate"] = np.where(df["Clicks_num"] > 0,
                                 df["Conversions_num"] / df["Clicks_num"],
                                 0)

df["cpa"] = np.where(df["Conversions_num"] > 0,
                     df["Cost_num"] / df["Conversions_num"],
                     np.nan)

df["cpc"] = np.where(df["Clicks_num"] > 0,
                     df["Cost_num"] / df["Clicks_num"],
                     np.nan)

# 3. Исследовательский анализ (EDA)

print("Общие суммарные метрики:")
summary = {
    "Total Impressions": df["Impr"].sum(),
    "Total Clicks": df["Clicks_num"].sum(),
    "Total Conversions": df["Conversions_num"].sum(),
    "Total Cost": df["Cost_num"].sum()
}
summary

# Средние метрики по всему датафрейму
overall_ctr = df["ctr_calc"].mean()
overall_cr = df["conversion_rate"].mean()
overall_cpa = df["cpa"].mean()

print("Средний CTR:         ", round(overall_ctr, 4))
print("Средний Conversion Rate:", round(overall_cr, 4))
print("Средний CPA:         ", round(overall_cpa, 2))

# 3.1 Сравнение Search vs Video

group_type = df.groupby("Campaign type").agg(
    impressions=("Impr", "sum"),
    clicks=("Clicks_num", "sum"),
    conversions=("Conversions_num", "sum"),
    cost=("Cost_num", "sum")
).reset_index()

group_type["ctr"] = group_type["clicks"] / group_type["impressions"]
group_type["conversion_rate"] = np.where(group_type["clicks"] > 0,
                                         group_type["conversions"] / group_type["clicks"],
                                         0)
group_type["cpa"] = np.where(group_type["conversions"] > 0,
                             group_type["cost"] / group_type["conversions"],
                             np.nan)

print("Сводка по типу кампании (Search vs Video):")
display(group_type)

# Пример графика CPA по типу кампании

group_type.plot(x="Campaign type", y="cpa", kind="bar", legend=False)
plt.ylabel("CPA (Cost per Conversion)")
plt.title("CPA by Campaign type")
plt.show()


# 3.2 Анализ по аудиториям

group_audience = df.groupby("Audience").agg(
    impressions=("Impr", "sum"),
    clicks=("Clicks_num", "sum"),
    conversions=("Conversions_num", "sum"),
    cost=("Cost_num", "sum")
).reset_index()

group_audience["ctr"] = group_audience["clicks"] / group_audience["impressions"]
group_audience["conversion_rate"] = np.where(group_audience["clicks"] > 0,
                                             group_audience["conversions"] / group_audience["clicks"],
                                             0)
group_audience["cpa"] = np.where(group_audience["conversions"] > 0,
                                 group_audience["cost"] / group_audience["conversions"],
                                 np.nan)

print("Сводка по аудиториям:")

display(group_audience.sort_values("conversions", ascending=False))


# 3.3 Анализ по устройствам

group_device = df.groupby("Device").agg(
    impressions=("Impr", "sum"),
    clicks=("Clicks_num", "sum"),
    conversions=("Conversions_num", "sum"),
    cost=("Cost_num", "sum")
).reset_index()

group_device["ctr"] = group_device["clicks"] / group_device["impressions"]
group_device["conversion_rate"] = np.where(group_device["clicks"] > 0,
                                           group_device["conversions"] / group_device["clicks"],
                                           0)
group_device["cpa"] = np.where(group_device["conversions"] > 0,
                               group_device["cost"] / group_device["conversions"],
                               np.nan)

print("Сводка по устройствам:")
display(group_device)

# 3.4 Анализ по ad groups (топ по конверсиям)

group_ad = df.groupby("Ad group").agg(
    impressions=("Impr", "sum"),
    clicks=("Clicks_num", "sum"),
    conversions=("Conversions_num", "sum"),
    cost=("Cost_num", "sum")
).reset_index()

group_ad["ctr"] = group_ad["clicks"] / group_ad["impressions"]
group_ad["conversion_rate"] = np.where(group_ad["clicks"] > 0,
                                       group_ad["conversions"] / group_ad["clicks"],
                                       0)
group_ad["cpa"] = np.where(group_ad["conversions"] > 0,
                           group_ad["cost"] / group_ad["conversions"],
                           np.nan)

print("Top ad groups по конверсиям:")
display(group_ad.sort_values("conversions", ascending=False).head(10))


# 4. Кластеризация рекламных связок (по CTR, CR, CPA и объёму трафика)

cluster_df = df.copy()

# Берём признаки для кластеризации
cluster_features = ["ctr_calc", "conversion_rate", "cpa", "Impr", "Clicks_num"]

# Заменяем NaN (например, в cpa) на медианы
cluster_df[cluster_features] = cluster_df[cluster_features].fillna(cluster_df[cluster_features].median())

from sklearn.preprocessing import StandardScaler
scaler_cl = StandardScaler()
cluster_scaled = scaler_cl.fit_transform(cluster_df[cluster_features])

# Выбираем число кластеров (например, 3)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(cluster_scaled)

df["cluster"] = cluster_labels

print("Распределение строк по кластерам:")
df["cluster"].value_counts()

# Сводка по кластерам: средние метрики
cluster_summary = df.groupby("cluster").agg(
    impressions=("Impr", "mean"),
    clicks=("Clicks_num", "mean"),
    conversions=("Conversions_num", "mean"),
    cost=("Cost_num", "mean"),
    ctr=("ctr_calc", "mean"),
    conversion_rate=("conversion_rate", "mean"),
    cpa=("cpa", "mean")
).reset_index()

print("Характеристики кластеров:")
display(cluster_summary)

# Пример: распределение аудиторий внутри кластеров

cluster_audience = df.groupby(["cluster", "Audience"]).size().reset_index(name="rows_count")
print("Распределение аудиторий по кластерам:")
display(cluster_audience)


# 5. Подготовка данных для моделирования Классификация + регрессия

# Для моделей берём только строки с ненулевыми кликами
model_df = df[(df["Clicks_num"] > 0) & (df["Impr"] > 0)].copy()

# Целевая переменная для классификации:
# effective = 1, если conversion_rate >= медианы; иначе 0
median_cr = model_df["conversion_rate"].median()
model_df["effective"] = (model_df["conversion_rate"] >= median_cr).astype(int)

print("Медиана Conversion Rate:", round(median_cr, 4))
print("Распределение классов effective:")
display(model_df["effective"].value_counts())

# Формируем признаки

features_cat = ["Campaign type", "Audience", "Device"]
features_num = ["Impr", "Clicks_num", "Cost_num", "ctr_calc", "conversion_rate", "cpa"]

X = model_df[features_cat + features_num].copy()
y_class = model_df["effective"]

# Заполняем пропуски в числовых признаках
X[features_num] = X[features_num].fillna(X[features_num].median())

# One-Hot Encoding категориальных признаков
X = pd.get_dummies(X, columns=features_cat, drop_first=True)

print("Финальный набор признаков для моделей:")
X.head()


# 6. Классификация: эффективная vs неэффективная связка

# Тренировочная и тестовая выборки для классификации
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X, y_class, test_size=0.3, random_state=42, stratify=y_class
)

# Масштабируем числовые признаки
scaler_c = StandardScaler()
X_train_num_scaled = scaler_c.fit_transform(X_train_c[features_num])
X_test_num_scaled = scaler_c.transform(X_test_c[features_num])

# Объединяем: сначала числовые (скейленные), затем категориальные (one-hot)
X_train_c_full = np.concatenate(
    [X_train_num_scaled, X_train_c.drop(columns=features_num).values],
    axis=1
)
X_test_c_full = np.concatenate(
    [X_test_num_scaled, X_test_c.drop(columns=features_num).values],
    axis=1
)

# 6.1 Logistic Regression

log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_c_full, y_train_c)

y_pred_lr = log_reg.predict(X_test_c_full)
y_proba_lr = log_reg.predict_proba(X_test_c_full)[:, 1]

acc_lr = accuracy_score(y_test_c, y_pred_lr)
f1_lr = f1_score(y_test_c, y_pred_lr)
roc_lr = roc_auc_score(y_test_c, y_proba_lr)

print("Logistic Regression:")
print("Accuracy:", round(acc_lr, 3))
print("F1:", round(f1_lr, 3))
print("ROC-AUC:", round(roc_lr, 3))
print("\nClassification report:\n", classification_report(y_test_c, y_pred_lr))

# 6.2 Random Forest Classifier

rf_clf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42
)
rf_clf.fit(X_train_c, y_train_c)

y_pred_rf = rf_clf.predict(X_test_c)
y_proba_rf = rf_clf.predict_proba(X_test_c)[:, 1]

acc_rf = accuracy_score(y_test_c, y_pred_rf)
f1_rf = f1_score(y_test_c, y_pred_rf)
roc_rf = roc_auc_score(y_test_c, y_proba_rf)

print("Random Forest Classifier:")
print("Accuracy:", round(acc_rf, 3))
print("F1:", round(f1_rf, 3))
print("ROC-AUC:", round(roc_rf, 3))
print("\nClassification report:\n", classification_report(y_test_c, y_pred_rf))

# Важность признаков в Random Forest

feature_importances = pd.Series(rf_clf.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False).head(15)


# 7. Регрессия: прогноз Conversion Rate

# Для регрессии берём те же признаки, но целевая — conversion_rate
y_reg = model_df["conversion_rate"]
X_reg = model_df[features_cat + features_num].copy()

# 1) Удаляем строки, где целевая переменная NaN
mask_target = y_reg.notna()
X_reg = X_reg[mask_target]
y_reg = y_reg[mask_target]

# 2) Заполняем пропуски в числовых признаках (Impr, Clicks_num, Cost_num, ctr_calc, conversion_rate, cpa)
X_reg[features_num] = X_reg[features_num].fillna(X_reg[features_num].median())

# 3) One-hot для категориальных признаков
X_reg = pd.get_dummies(X_reg, columns=features_cat, drop_first=True)

# 4) На всякий случай: ещё раз убедимся, что в X_reg больше нет NaN
print("Есть ли NaN в X_reg:", X_reg.isna().any().any())

# 5) Делим на train/test
from sklearn.model_selection import train_test_split

Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_reg, y_reg, test_size=0.3, random_state=42
)

# 6) Масштабируем признаки
from sklearn.preprocessing import StandardScaler

scaler_r = StandardScaler()
Xr_train_scaled = scaler_r.fit_transform(Xr_train)
Xr_test_scaled = scaler_r.transform(Xr_test)

# 7.1 Linear Regression

lin_reg = LinearRegression()
lin_reg.fit(Xr_train_scaled, yr_train)

yr_pred_lr = lin_reg.predict(Xr_test_scaled)

mae_lr = mean_absolute_error(yr_test, yr_pred_lr)
rmse_lr = np.sqrt(mean_squared_error(yr_test, yr_pred_lr))
r2_lr = r2_score(yr_test, yr_pred_lr)

print("Linear Regression (CR prediction):")
print("MAE:", round(mae_lr, 4))
print("RMSE:", round(rmse_lr, 4))
print("R^2:", round(r2_lr, 3))

# 7.2 Random Forest Regressor

rf_reg = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)
rf_reg.fit(Xr_train, yr_train)

yr_pred_rf = rf_reg.predict(Xr_test)

mae_rf = mean_absolute_error(yr_test, yr_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(yr_test, yr_pred_rf))
r2_rf = r2_score(yr_test, yr_pred_rf)

print("Random Forest Regressor (CR prediction):")
print("MAE:", round(mae_rf, 4))
print("RMSE:", round(rmse_rf, 4))
print("R^2:", round(r2_rf, 3))


# 8. Интерпретация: сегменты и рекомендации (заготовка для аналитических выводов)


# Пример: доля эффективных связок по аудиториям
eff_by_audience = model_df.groupby("Audience")["effective"].mean().reset_index()
eff_by_audience = eff_by_audience.rename(columns={"effective": "share_effective"})
print("Доля эффективных связок по аудиториям:")
display(eff_by_audience.sort_values("share_effective", ascending=False))

# Пример: средний CR и CPA по кластерам
print("Средний CR и CPA по кластерам:")
display(df.groupby("cluster")[["conversion_rate", "cpa"]].mean())

```
